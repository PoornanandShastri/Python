### Type Casting

```python
# In Python, we can convert one type of variable to another type. 
# This conversion is called type casting or type conversion.

# Python performs the following two types of casting.

# 1. Implicit casting:- The Python interpreter automatically performs an implicit 
# Type conversion, which avoids loss of data.

# 2. Explicit casting:- The explicit type conversion is performed by the user 
# using built-in functions.


# In type casting, data loss may occur because we enforce the object to a 
# specific data type.
```

### Int Type Conversion

```python
# In int type conversion, we use the int() function to convert variables of other 
# types to int type. Variable can be of any type such as float, string, bool.

# While performing int type conversion, we need to remember the following points.

# 1. When converting string type to int type, a string must contain integral value 
# only and should be base-10.

# 2. We can convert any type to int type, but we cannot perform complex to int type.


# 1. Type Cast float to int

floatVar1 = 4.5
print(floatVar1)
print(type(floatVar1))

intVar1 = int(floatVar1)
print(intVar1)
print(type(intVar1))


# 2. Type Cast Bool to Int

boolVar1 = True
boolVar2 = False

print(boolVar1)
print(boolVar2)

print(type(boolVar1))
print(type(boolVar2))

intVar2 = int(boolVar1)
intVar3 = int(boolVar2)

print(intVar2) # 1 means True
print(intVar3) # 0 means False

print(type(intVar2))
print(type(intVar3))


# 3. Type Cast Str to Int

strVar1 = "34"
print(strVar1)
print(type(strVar1))

intVar4 = int(strVar1)
print(intVar4)
print(type(intVar4))

# When converting string type to int type, a string must contain integral value only 
# and should be base-10. If you try to convert it will give error


strVar2 = "Vivek"
print(strVar2)
print(type(strVar2))

intVar5 = int(strVar2)
print(intVar5)
print(type(intVar5))
```

### Float Type Conversion

```python
# In float type conversion we use a built-in function  float() . 
# This function converts variables of other types to float types.


# 1. Type Cast int to float

intVar1 = 4
print(intVar1)
print(type(intVar1))

floatVar1 = float(intVar1)
print(floatVar1)
print(type(floatVar1))


# 2. Type Cast Bool to float

boolVar3 = True
boolVar4 = False
print(boolVar3)
print(boolVar4)
print(type(boolVar3))
print(type(boolVar4))

floatVar2 = float(boolVar3)
floatVar3 = float(boolVar4)
print(floatVar2)
print(floatVar3)
print(type(floatVar2))
print(type(floatVar3))


# 3. Type Cast str to float

strVar3 = "48"
print(strVar3)
print(type(strVar3))

floatVar4 = float(strVar3)
print(floatVar4)
print(type(floatVar4))

# Same as int type conversion string must contain decimal value otherwise it'll give 
# error

# While performing float type conversion, we need to remember some points.

# We can convert any type to float type, but we cannot cast complex to float type.

# While converting string type to float type, a string must contain an integer/decimal 
# value of base-10.
```

### Complex Type Conversion

```python
# In complex type conversion, we use the built-in function complex() to convert 
# values from other types to the complex type. Value can be any type including of  
# int, float, bool, str.


# The complex function has the following two forms for conversion.

# 1. complex(x): To convert a value x into a complex type. In this form, the real 
# value is  x, and the imaginary value is 0.

# 2. complex(x, y): To convert the value x and y into a complex type. In this form, 
# the real value is x, and the imaginary is y.


# 1. Type Cast int to Complex

intVar5 = 8
print(intVar5)
print(type(intVar5))

complexVar1 = complex(intVar5)
print(complexVar1)
print(type(complexVar1))

# here we have pass only 1 value so it will take that as real part and imagionary 
# value will be 0


intVar6 = 8
intVar7 = 2
print(intVar6)
print(intVar7)
print(type(intVar6))
print(type(intVar7))

complexVar2 = complex(intVar6, intVar7)
print(complexVar2)
print(type(complexVar2))


# 2. Type Cast float to Complex

floatVar5 = 7.5
print(floatVar5)
print(type(floatVar5))

complexVar3 = complex(floatVar5)
print(complexVar3)
print(type(complexVar3))

# here we have pass only 1 value so it will take that as real part and imagionary 
# value will be 0


floatVar6 = 3.5
floatVar7 = 2.6
print(floatVar6)
print(floatVar7)
print(type(floatVar6))
print(type(floatVar7))

complexVar4 = complex(floatVar6, floatVar7)
print(complexVar4)
print(type(complexVar4))


# 3. Type Cast Bool to Complex

boolVar5 = True
boolVar6 = False
print(boolVar5)
print(boolVar6)
print(type(boolVar5))
print(type(boolVar6))

complexVar5 = complex(boolVar5)
complexVar6 = complex(boolVar5,boolVar6)
print(complexVar5)
print(complexVar6)
print(type(complexVar5))
print(type(complexVar6))


# 4. Type Cast Str to Complex
# We can not perform this type conversion

```

### Bool Type conversion

```python
# We use the built-in function bool() to convert values of other types to bool types. 
# This function returns two values, either True or False.

# We can convert any type of values to bool type, and the output for all values will 
# be True, Except 0, which is False.

# If you convert an empty string to a boolean it will be converted to boolean False.

# The bool True is 1 and False is 0. Every non-zero value is treated as True.


# 1. Type Cast int to Bool

intVar8 = 8 # any non zero value wil be true
intVar9 = 0 # zero always will be false
print(intVar8)
print(intVar9)
print(type(intVar8))
print(type(intVar9))

boolVar7 = bool(intVar8)
boolVar8 = bool(intVar9)
print(boolVar7)
print(boolVar8)
print(type(boolVar7))
print(type(boolVar8))


# 2. Type Cast Float to Bool

floatVar8 = 5.8 # any non zero value wil be true
floatVar9 = 0.0 # zero always will be false
print(floatVar8)
print(floatVar9)
print(type(floatVar8))
print(type(floatVar9))

boolVar9 = bool(floatVar8)
boolVar10 = bool(floatVar9)
print(boolVar9)
print(boolVar10)
print(type(boolVar9))
print(type(boolVar10))


# 3. Type Cast Str to Bool

strVar6 = "False" 
strVar7 = "True" 
strVar8 = "456" 
strVar9 = "" 

print(strVar6)
print(strVar7)
print(strVar8)
print(strVar9)
print(type(strVar6))
print(type(strVar7))
print(type(strVar8))
print(type(strVar9))

boolVar11 = bool(strVar6)
boolVar12 = bool(strVar7)
boolVar13 = bool(strVar8)
boolVar14 = bool(strVar9)
print(boolVar11)
print(boolVar12)
print(boolVar13)
print(boolVar14)
print(type(boolVar11))
print(type(boolVar12))
print(type(boolVar13))
print(type(boolVar14))


# 4. Type Cast Complex to Bool

complexVar8 = 12 + 8j
complexVar9 = 0 + 0j 
print(complexVar8)
print(complexVar9)
print(type(complexVar8))
print(type(complexVar9))

boolVar15 = bool(complexVar8)
boolVar16 = bool(complexVar9)
print(boolVar15)
print(boolVar16)
print(type(boolVar15))
print(type(boolVar16))
```

### String Type Conversion

```python
# In str type conversion, we use the built-in function str() to convert converts 
# variables of other types to a string type. This function returns the string 
# type of object (value).

# 1. Type Cast int to str

intVar10 = 40
print(intVar10)
print(type(intVar10))

strVar10 = str(intVar10)
print(strVar10)
print(type(strVar10))


# 2. Type Cast float to str

floatVar10 = 22.87
print(floatVar10)
print(type(floatVar10))

strVar11 = str(floatVar10)
print(strVar11)
print(type(strVar11))


# 3. Type Cast complex to str

complexVar10 = 4+5j
print(complexVar10)
print(type(complexVar10))

strVar12 = str(complexVar10)
print(strVar12)
print(type(strVar12))


# 3. Type Cast Bool to str

boolVar17 = True
boolVar18 = False

print(boolVar17)
print(boolVar18)

print(type(boolVar17))
print(type(boolVar18))

strVar13 = str(boolVar17)
strVar14 = str(boolVar18)

print(strVar13)
print(strVar14)

print(type(strVar13))
print(type(strVar14))

```
